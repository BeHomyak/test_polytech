```
class box:
  def __init__(self, data=None):
    self.data=data
    self.lenght=0
    self.next=None

class boxlist:
  def __init__(self):
    self.head=None

  def boxprint (self):
    p_box=self.head
    while p_box is not None:
      print(p_box.data)
      p_box=p_box.next

  def create (self, newdata):
    newbox=box(newdata)
    if self.head is None:
        self.head=newbox
        return
    last=self.head
    while (last.next):
        last=last.next
    last.next=newbox

  def remove (self, position):
    r_box=self.head
    i=1
    if r_box is not None:
      if position == 1:
        self.head=r_box.next
        r_box=None
        return
      else:
        while r_box is not None:
          if i == position:
            break
          last=r_box
          r_box=r_box.next
          i+=1
        last.next=r_box.next
          
  def insert(self, position, newdata):
    newbox=box(newdata)
    i_box=self.head
    i=1
    if position == 1:
      last=i_box
      self.head=newbox
      newbox.next=last
    else:
      while i_box is not None:
        if i == position:
          last.next=newbox
          newbox.next=i_box
          break
        last=i_box
        i_box=i_box.next
        i+=1
      if (i_box is None) and (i == position):
        last.next=newbox

  def search (self, data_s):
    s_box=self.head
    k=0
    t=0
    while s_box is not None:
      k+=1
      if s_box != None:
        if s_box.data == data_s:
          t=1
          print('Искомый элемент',data_s,'находится на месте',k)
      s_box=s_box.next
      if (s_box == None) and (t==0):
        print('Элемент',data_s,'не найден')

Box=boxlist()
Box.create(10)
Box.create(11)
Box.create(12)
Box.create(13)
Box.create(14)
Box.create(15)
Box.boxprint()

Box.search(16)

Box.remove(4)

Box.insert(1, 1)
Box.insert(5, 2)
Box.insert(8, 3)

Box.boxprint()
```
